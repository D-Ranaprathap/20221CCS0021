```python
#heat maps
data=np.random.rand(13,16)
plt.imshow(data,cmap="magma")
plt.title=("2d heat map")
plt.colorbar()
plt.show()
```

```python
#heat maps using seaborn
data=np.random.randint(low=14,high=100,size=(5,5))
hm=sns.heatmap(data=data,annot=True)
plt.show()
```
