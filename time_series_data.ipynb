```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2021-1-1","2021-6-12","2021-7-31","2022-1-10","2022-1-15","2022-11-1","2022-12-16","2023-1-5","2023-7-1","2023-10-3",]),
    "value":[10,15,18,19,20,22,25,29,30,32]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o')
plt.xlabel("date")
plt.ylabel("value")
plt.title("time series with trend")
plt.grid(True)
plt.show()
```
